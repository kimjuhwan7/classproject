```python
import numpy as np
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x= [1,2,3,4]
y=[2,4,6,8]

x2 = np.array([1,2,3,4])
y2 = np.array([4,1,3,1])
plt.plot(x,y, color = 'red',linestyle='dotted',marker = '^')
plt.plot(x2,y2, color = 'green',linestyle='dashed',marker = 'o')
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x= [1,2,3,4]
y=[2,4,6,8]

x2=[2,2,6,5]
y2=[4,4,6,6]

axes.bar(x,y,color='yellow')
axes.bar(x2,y2,color='green')
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)
axes2 = axes.twinx()
x= [1,2,3,4]
y=[2,4,6,8]

x2=[1,2,3,4]
y2=[4,4,6,6]

axes.bar(x,y,color='yellow')
axes2.plot(x2,y2,color='green',label='plot')

axes.legend()
axes2.legend(loc=1)
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)
axes2 = axes.twinx()
x= [1,2,3,4]
y=[2,4,6,8]

x2=[1,2,3,4]
y2=[4,6,1,9]

axes.scatter(x,y,color='yellow')
axes.scatter(x2,y2,color='g')

plt.show()
```

```python
# wordcloud
import wordcloud
text = '파이썬 파이썬 파이썬 시각화 데이터분석 오류 데이터'
font_path= '/Windows/Fonts/H2GTRE.TTF'
# 시각화 이미지 설정: 한글폰트, 배경색, 
wc = wordcloud.WordCloud(font_path=font_path, background_color='pink',stopwords=['오류','데이터'])
wc.generate_from_text(text)
wc.to_image()
```
