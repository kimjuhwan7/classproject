```python


data={}

for i in range(1,len(result)):
    data[result.iloc[i,0]] =result.iloc[1,range(1,len(result.columns),6)]
    
```

```python
print(data)
```

```python

plt.ylim([140000,161000])

loc = '서울특별시 용산구 (1117000000)'

for i in range(len(data[loc])):
    #str -> ',' 를 제거 -> 문자를 숫자 타입으로 변경
    data[loc][i] = int(data[loc][i].replace(',',''))
    plt.bar(data[loc].index[i].split('년_')[0],data[loc][i])

plt.show()
```
